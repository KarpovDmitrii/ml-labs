```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer

#df = df.iloc[:40]

def list_to_str(x):
    s=""
    for i in x:
        s+=i+" "
    return s

def lemmatization(x):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(x)
    data_lematized = [word.lemma_ for word in doc]
    return(list_to_str(data_lematized))

# вот до сюда не использую, тут я просто пытался сделал лемматизацию 
# добавляю в лист тексты из строчек
a=[]
for i in range(df.shape[0]):
    a.append(df["v2"][i])
# применяю bag- of words
vect = CountVectorizer(min_df=5, stop_words='english')
bow = vect.fit_transform(a)
bow.shape
```

Я тут циферки прикинул, за одну секунду обрабатывается две строчки, у меня 5000 строк, это 2 часа 40 минут. Сегодня будет без лемматизации

```python
# преобразую векторы из циферок в датафрейм
words = pd.DataFrame(bow.A, columns=vect.get_feature_names_out())
```

```python
words
```

```python
target=df["v1"]
target
```

```python
data = pd.concat([target, words], axis = 1,ignore_index = True)
data.shape
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
from sklearn.decomposition import PCA
# унижаю размерность предикторов
pca = PCA(n_components=5)
X_pca = pca.fit_transform(words, target)

print(words.shape)
print(X_pca.shape)
```

```python
# *** ИСПОЛЬЗОВАТЬ САМОСТОЯТЕЛЬНО РЕАЛИЗОВАННЫЕ МОДЕЛИ ИЗ ПРЕДЫДУЩИХ ЛР
from math import sqrt
from scipy.stats import mode
import numpy as np

class KNNMethod:
    def __init__(self, x_train, y_train, k_neighbors):
        self.x_train=x_train
        self.y_train=y_train.array
        self.k_neighbors = k_neighbors

    def predict(self, x_test):
        x_test=x_test
        classes=[]
        for test_index, test_predictors in enumerate(x_test):
            distances  = []
            for train_index, train_predictors in enumerate(self.x_train):
                distance=0
                for i in range(len(test_predictors)):
                    distance += (test_predictors[i] - train_predictors[i]) ** 2
                distances.append(
                    [sqrt(distance), self.y_train[train_index]]
                )
            distances.sort(key=lambda x: x[0])
            neighbours = distances[:self.k_neighbors]
            classes.append(mode(np.array(neighbours)[:, 1]).mode)
        return classes
```

```python
#провожу классификацию с помощью моего кнн

X_train, X_test, y_train, y_test = train_test_split(X_pca, target, test_size=0.33)
bag = KNNMethod(X_train, y_train, 9)
print(classification_report(y_test, bag.predict(X_test)))
```

```python
#провожу классификацию с помощью бэггинга

X_train, X_test, y_train, y_test = train_test_split(X_pca, target, test_size=0.33)
bag = BaggingClassifier().fit(X_train, y_train)
print(classification_report(y_test, bag.predict(X_test)))
```

## LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
```

```python
n_samples = 2000
n_features = 1000
n_topics = 10
n_top_words = 20
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
```

```python
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=20,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0).fit(data)
```

```python
# матрица темы-слова
lda.components_.shape
```

```python
# матрица документы-темы
lda.transform(data).shape
```
